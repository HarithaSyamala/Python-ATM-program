```python
print("Welcome to Northern Frost\n\nInsert Your Card")

password=1234
balance=100
choice=0

pin=int(input("Enter Your Four Digit Pin\n"))
if pin==password:
  while choice != 4:

    print("\n\n**** Menu ****")
    print("1 == Dispalay balance")
    print("2 == Wthdraw funds")
    print("3 == Deposit funds")
    print("4 == Return card")

    choice=int(input("\nEnter your option:\n "))

    if choice==1:
      print("Display balance = £", balance)

    elif choice==2:
      wit=int(input("Enter the funds to withdraw: £"))
      balance -= wit
      print("\nwithdrawn funds: £", wit)
      print("balance = £", balance)

    elif choice==3:
      dep=int(input("Enter your deposit: £"))
      balance += dep
      print("\ndeposited funds: £", dep)
      print("balance = £", balance)

    elif choice==4:
      print("\n Return card!! Goodbye")

    else:
      print("\nInvalid Entry!!")

else:
  print("Incorrect pin!! Try again")




```
